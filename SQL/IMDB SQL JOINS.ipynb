```python
def sql(query,db="imdb"):
    try:
        mydb = connection.connect(host="localhost", database = db,user="root", passwd="mysql",use_pure=True)
    # check if the connection is established
        #print(mydb.is_connected())
        result_dataFrame = pd.read_sql(query,mydb)
        return result_dataFrame
        mydb.close()
    except Exception as e:
        mydb.close()
        print(str(e))
```

# JOIN (Inner Join)

```python
sql("SELECT m.name, g.genre from movies m  JOIN movies_genres g ON m.id=g.movie_id LIMIT 20;") ## 
```

# Left Join 

```python
sql("SELECT m.name, g.genre from movies m  LEFT JOIN movies_genres g ON m.id=g.movie_id LIMIT 20;") ## this will show all the movies even if genre is not there
```

#  Right Join

```python
sql("SELECT m.name, g.genre from movies m  RIGHT JOIN movies_genres g ON m.id=g.movie_id LIMIT 20;") ## it will show all the movies which has genre
```

# FULL OUTER JOIN 

## statement of left join union statement of right join

```python
sql(r"SELECT m.name, g.genre from movies m  LEFT JOIN movies_genres g ON m.id=g.movie_id UNION SELECT m.name, g.genre from movies m  RIGHT JOIN movies_genres g ON m.id=g.movie_id LIMIT 20;")
```
