```python
import mysql.connector as connection ##
import pandas as pd
try:
    mydb = connection.connect(host="localhost", user="root", passwd="mysql",use_pure=True) ## passwd add your password
    # check if the connection is established
    print(mydb.is_connected())
    mydb.close()
    
except Exception as e:
    print(str(e))
```

```python
def sql(query,db="practise_data"):
    try:
        mydb = connection.connect(host="localhost", database = db,user="root", passwd="mysql",use_pure=True) ## connection
    # check if the connection is established
        #print(mydb.is_connected())
        result_dataFrame = pd.read_sql(query,mydb) ## pandas can read sql
        return result_dataFrame
        mydb.close()
    except Exception as e:
        mydb.close()
        print(str(e))
```
